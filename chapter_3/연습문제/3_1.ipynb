## 풀이
평균이 200이고, 표준편차가 20인 정규분포를 따르는 데이터 집합이 있을때, 데이터의 60%가 속할 그 범위를 찾는 문제이다.<br>
위에서 정규분포를 따른다고 했으므로, 평균을 중심으로 대칭이 되는 종모양의 정규분포를 고려해볼 수 있다.<br>
이때 표준편차를 고려하여 표준편차의 배수만큼 변화량이 조절되어 평균에서 떨어지는 z-점수를 구할 수 있다<br>
이를 구하기 위해,상위 20%와 하위 20%를 소거하여 남은 범위를 구한다.

```python
from scipy.stats import norm
```

```python
#평균과 표준편차 설정
mean = 200
std_dev = 20
```

```python
# 유의수준 설정
alpha = 1-0.6
```

```python
#상위 20%에 해당하는 z-점수 찾기(양쪽 꼬리에서 각각 20%씩 제외하므로, 양측 총합 40%에 해당하는 지점 찾기)
z_score_upper = norm.ppf(1-alpha/2)
```

```python
#범위의 하한과 상한 계산
lower_bound = mean - (z_score_upper*std_dev)
upper_bound = mean + (z_score_upper*std_dev)
```

```python
print('lower_bound: ', lower_bound)
print('upper_bound: ', upper_bound)
print('z_score_upper: ', z_score_upper)
```

## 만약에...
**만약 평균이 0인 정규분포라면?**

```python
mean = 0
std_dev = 20
lower_bound2 = mean - (z_score_upper*std_dev)
upper_bound2 = mean + (z_score_upper*std_dev)
print('lower_bound: ', lower_bound2)
print('upper_bound: ', upper_bound2)
print('z_score_upper: ', z_score_upper)
```
