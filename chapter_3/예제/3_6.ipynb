```python
import numpy as np
from scipy import stats

x = [1.08, 1.03, 0.96, 0.95, 1.04, 1.01, 0.98, 0.99, 1.05, 1.08, 0.97, 1.00, 0.98, 1.01]
#   

"""
x_bar: 
S_x: 
"""

x_bar = np.mean(x)
S_x = np.std(x, ddof=1) #    ddof 1 

print(f"(x_bar: {x_bar}, S_x: {S_x})")
```

```python
"""
n: (Sample Size)
v: (Degree Of Freedom)
alpha: (level of significance)
"""

n = 14
v = n-1 
alpha = 1-0.95
t_critical_value = stats.t.ppf(1-alpha/2, v)
print(f"{t_critical_value:.2f}")
```

```python
"""
moe: (margin of error)
lb: (upper bound)
ub: (lower bound)
"""
moe = t_critical_value * (S_x / np.sqrt(n))

lb = x_bar - moe
ub = x_bar + moe

print(f"{x_bar:.3f} - {moe:.3f} < meu < {x_bar:.3f} + {moe:.3f}")
print(f"{lb:.3f} < meu < {ub:.3f}")
```
