문제a)
500Hz의 정현파 신호가 750Hz로 샘플링된다고 가정하라. 이산시간신호를, 그리고 신호의 명백하 주파수를 결정하라.

문제b)
만약 200Hz의 구성요소가 (a)부분의 신호에 존재한다면, 감지될 것인가? 설명하라

문제c)
만약 375Hz의 구성요소가 (a)부분의 신호에 존재한다면, 감지될 것인가? 설명하라

4.12a) <br>
이 경우 나이퀴스트 주파수는 샘플링주파수(750)의 절반인 375Hz이다. 500Hz의 신호는 나이퀴스트 주파수를 초과하므로, 엘리어싱이 발생하게 될 것이다. 엘리어싱을 거쳐 샘플링된 신호의 주파수는 샘플링 주파수에서 실제 주파수를 뺸 값, 750 - 500 = 250(Hz)가 될 것이다.

4.12b) <br>
200Hz의 구성요소가 있었다면, <br>
이 요소는 나이퀴스트 주파수 내에 있으므로 정확하게 감지될 것이다.

4.12c) <br>
375Hz의 구성요소가 있었다면, <br>
이론상으로 정확하게 감지될수는 있으나, 나이퀴스트 주파수와 딱 일치하기 때문에 샘플링이 제대로 되지 않을수도 있다. 이는 신호의 위상에 따라 다른결과가 나타날 수 있기 때문이다.
