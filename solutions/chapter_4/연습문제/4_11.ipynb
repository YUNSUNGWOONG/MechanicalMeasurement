4.11c) <br>
200Hz로 샘플링할 때, 실제 주파수가 나이퀴스트 주파수인 100Hz를 넘는 모든 주파수는 엘리어싱의 영향을 받는다. 이 경우 150Hz는 나이퀴스트 주파수보다 50Hz 높으므로, 샘플링된 신호는 200 - 150 = 50Hz의 주파수를 가진 것처럼 나타날 것이다. 이것은 200Hz 샘플링주파수로 인해 실제 주파수보다 낮은 주파수로 측정된다는 것을 의미한다.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the signal frequency and sampling rate
signal_frequency = 150  # in Hz
sampling_rate = 200  # in Hz

# Generate the signal over a period of time
t = np.linspace(0, 1, 1000, endpoint=False)  # 1 second duration, high resolution for continuous signal
cosine_wave = np.cos(2 * np.pi * signal_frequency * t)

# Calculate the sampling points
sampling_points = np.arange(0, 1, 1/sampling_rate)
sampled_wave = np.cos(2 * np.pi * signal_frequency * sampling_points)

# Plot the signal
plt.figure(figsize=(10, 4))
plt.plot(t, cosine_wave, label='Continuous 150Hz Cosine Wave')
plt.scatter(sampling_points, sampled_wave, color='red', label='Sampled at 200Hz')
plt.title('Sampling 150Hz Cosine Wave at 200Hz')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()

# Determine the apparent frequency due to aliasing
nyquist_frequency = sampling_rate / 2
apparent_frequency = nyquist_frequency - (signal_frequency - nyquist_frequency)

apparent_frequency
```
