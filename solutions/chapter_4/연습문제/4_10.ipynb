```python
import numpy as np
import matplotlib.pyplot as plt

#      
actual_frequency = 1000  #    1.5kHz
sampling_frequency = 2000  #   1kHz
sampling_period = 1 / sampling_frequency  #  
time = np.arange(0, 2, sampling_period)  # 2   

#  
input_signal = np.sin(2 * np.pi * actual_frequency * time)
output_signal = np.sin(2 * np.pi * sampling_frequency * time)

#  
plt.figure(figsize=(18, 6))

plt.subplot(1,2,1)
plt.plot(time, input_signal, 'o-')
plt.title('Input')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(time, output_signal, 'o-')
plt.title('Output(Aliasing)')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.show()
```
