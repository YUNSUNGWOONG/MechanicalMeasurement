```python
# N만큼의 주파수로 샘플링되어 N/2개의 점들이 생성되었으므로 N/2가 나이퀴스트 주파수가 된다.
sampling_frequency = 4096
nyquist_frequency = sampling_frequency / 2
print(f'4.10a) {nyquist_frequency}')
```

```python
# 주파수해상도는 총 샘플링 기간에 대한 역수
# 주파수해상도 = (샘플링 주파수)/(포인트의 수)
number_of_points=2048

frequency_resolution = 4096 / 2048
print(f"4.10b) {frequency_resolution}")
```

4.10c) <br>
2048Hz의 나이퀴스트 주파수에서, 500Hz의 정수 배수인 조화성분은 2048Hz 이하에서 정확하게 측정될 수 있다. 예를 들어, 500Hz, 1000Hz, 1500Hz, 2000Hz의 조화성분은 정확히 측정될 것이다. 하지만 2048헤르츠 이상인 2500Hz 등 나이키스트 주파수를 초과하는 조화성분들은 엘리어싱으로 인해 고주파신호임에도 저주파 신호로 잘못 인식하여 출력하게 되는 문제가 발생한다

```python
import numpy as np
import matplotlib.pyplot as plt

# 실제 신호 주파수와 샘플링 주파수 설정
actual_frequency = 1000  # 실제 신호 주파수 1.5kHz
sampling_frequency = 2000  # 샘플링 주파수 1kHz
sampling_period = 1 / sampling_frequency  # 샘플링 주기
time = np.arange(0, 2, sampling_period)  # 2초 동안의 시간 벡터

# 신호 생성
input_signal = np.sin(2 * np.pi * actual_frequency * time)
output_signal = np.sin(2 * np.pi * sampling_frequency * time)

# 신호 플롯
plt.figure(figsize=(18, 6))

plt.subplot(1,2,1)
plt.plot(time, input_signal, 'o-')
plt.title('Input')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(time, output_signal, 'o-')
plt.title('Output(Aliasing)')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.show()
```
