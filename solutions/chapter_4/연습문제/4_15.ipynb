```python
from scipy.constants import pi
import sympy as sp

# Given values
f = 10  # frequency in Hz
m = 60.10e-3  # mass in kg (converted from grams)
m_uncertainty = 0.11e-3  # mass uncertainty in kg

# Beam stiffness formula k = (2*pi*f)**2 * m
k = (2 * pi * f)**2 * m

# Calculating the uncertainty in k
# We'll use the sympy package for symbolic calculation to find the partial derivative of k with respect to m
m_symbolic = sp.symbols('m')
k_symbolic = (2 * sp.pi * f)**2 * m_symbolic
k_prime = sp.diff(k_symbolic, m_symbolic)  # First derivative of k with respect to m

# The uncertainty in k is then given by multiplying this derivative with the uncertainty in m
k_uncertainty = k_prime.evalf(subs={m_symbolic: m}) * m_uncertainty

k, k_uncertainty
```
