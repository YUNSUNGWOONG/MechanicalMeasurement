```python
import sympy as sp
import numpy as np
w,f = sp.symbols('w, f')

#방정식 결정하기
f=w/(2*sp.pi)

#함수에 값 대입하기
f_val = f.subs({w: 15})

print(f"4.1a) {f_val:.3f}")
```

```python
# 심볼정의
t = sp.symbols('t')

# 상수 초기화
omega = 15
#pi = 3.14

# Given equation in terms of sine and cosine
y = 100 + 95 * sp.sin(omega * t) + 55 * sp.cos(omega * t)

# sin(x) = cos(x - pi/2)
# rewrite_only_cosine = 100 +  95*sp.cos(omega*t - pi/2) + 55 * sp.cos(omega*t)
rewrite_only_cosine_1= 100 + 95*sp.cos(omega*t-sp.pi/2) + 55*sp.cos(omega*t)

R=round(np.sqrt(95**2 + 55**2),2)
seta=round(np.arctan(95/55),2)
rewrite_only_cosine_2=100 + R*sp.cos(omega*t + seta)

print(f"4.1b) {rewrite_only_cosine_1.simplify()}")
print(f"   or  {rewrite_only_cosine_2.simplify()}")
```
