```python
import numpy as np
import matplotlib.pyplot as plt

# Define the fundamental frequency based on the period of the square wave in the figure
# The figure shows 3 periods over a range of 6, thus the period T = 2 and the fundamental frequency f = 1/T
T = 2 * np.pi
fundamental_freq = 1 / T

# Generate harmonics and their amplitudes
# Only odd harmonics (1, 3, 5, ...) contribute to the square wave
harmonics = np.arange(1, 20, 2)
amplitudes = 1 / harmonics  # The amplitude of the harmonics decreases as 1/n

# Create a stem plot for the frequency spectrum
plt.figure(figsize=(10, 5))
plt.stem(harmonics * fundamental_freq, amplitudes)
plt.title('Frequency Spectrum of the Square Wave')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the frequency and its harmonics for the square wave
# Since the square wave period is 2pi, the fundamental frequency (f1) is 1/(2pi)
fundamental_freq = 1/(2 * np.pi)
harmonics = np.arange(1, 50, 2)  # We consider the first 25 odd harmonics

# The amplitude of the nth harmonic of the square wave is 1/n
amplitudes = 1/harmonics

# Plot the frequency spectrum
plt.figure(figsize=(10, 4))
plt.stem(harmonics * fundamental_freq, amplitudes, basefmt=" ")
plt.title('Frequency Spectrum of a Square Wave')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
