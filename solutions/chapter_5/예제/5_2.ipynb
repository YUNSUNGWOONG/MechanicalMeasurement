```python
#      .
import numpy as np

#    
T0 = 75  #   (F)
T_inf = 300  #   (F)
tau = 6  #  ()
t = 10  #   ()

# 1   
T_t = T_inf + (T0 - T_inf) * np.exp(-t / tau)

print(f"   10   : {T_t:.2f} F")
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic')

#    
T0 = 75  #   (F)
T_inf = 300  #   (F)
tau = 6  #  ()
t_end = 20  #    ()

#   
t = np.linspace(0, t_end, 1000)

# 1   
T_t = T_inf + (T0 - T_inf) * np.exp(-t / tau)

# 
plt.figure(figsize=(10, 6))
plt.plot(t, T_t, label=' ', color='orange')
plt.axvline(x=10, color='r', linestyle='--', label='t = 10')
plt.title('    (300F 75F  )')
plt.xlabel(' ()')
plt.ylabel(' (F)')
plt.grid(True)
plt.legend()
plt.show()
```
