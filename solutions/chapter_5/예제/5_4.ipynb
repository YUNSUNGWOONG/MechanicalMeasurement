```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic')

#    
T_min = 75  #   (F)
T_max = 300  #   (F)
tau = 10  #  ()
T0 = 187.5  #    (F)
A = 112.5  #    (F)
T_period = 20  #    ()
omega = 2 * np.pi / T_period  #  (rad/s)

#   
t = np.linspace(0, 60, 1000)  # 60  

#   
T_in = T0 + A * np.cos(omega * t)

# 1    (   )
T_out = T0 + A / np.sqrt(1 + (omega * tau)**2) * np.cos(omega * t - np.arctan(omega * tau))

# 
plt.figure(figsize=(10, 6))
plt.plot(t, T_in, label='  ', color='blue')
plt.plot(t, T_out, label='   ', color='orange')
plt.title('     ')
plt.xlabel(' ()')
plt.ylabel(' (F)')
plt.grid(True)
plt.legend(loc="upper right")
plt.show()
```
