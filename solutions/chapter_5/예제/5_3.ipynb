```python
import numpy as np

# 초기조건:공식풀이의 온도표기를 고려해서 나도 P를 쓰기로 했다.
P_inf = 75
P_A = 300
t = 10
tau = 6

# 1차 시스템의 응답 계산
P = 75 + (P_A - P_inf) * np.exp(-t/tau)
print(f"과정이 시작된 후 10초가 지났을 때의 온도: {P:.2f} °F")
```
