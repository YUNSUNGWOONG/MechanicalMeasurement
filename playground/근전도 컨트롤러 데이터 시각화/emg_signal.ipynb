```python
#60hz제거
def filter(signal_data):
    sampling_freq=200
    nyquist_freq = 0.5 * sampling_freq
    nyquist_freq = 0.5 * sampling_freq
    low_cutoff = 59  
    high_cutoff = 61  
    band_stop_filter = signal.butter(N=4, Wn=[low_cutoff / nyquist_freq, high_cutoff / nyquist_freq], btype='bandstop', output='sos')
    filtered_signal = signal.sosfilt(band_stop_filter, signal_data)
    fft_result = np.fft.fft(filtered_signal)
    freq = np.fft.fftfreq(len(filtered_signal), d=1/sampling_freq)
    return (filtered_signal, fft_result, freq)

(filtered_signal_a, fft_result_a,freq_a) = filter(data['a'])
(filtered_signal_b, fft_result_b,freq_b) = filter(data['b'])
(filtered_signal_c, fft_result_c,freq_c) = filter(data['c'])
(filtered_signal_d, fft_result_d,freq_d) = filter(data['d'])
(filtered_signal_e, fft_result_e,freq_e) = filter(data['e'])
(filtered_signal_f, fft_result_f,freq_f) = filter(data['f'])
(filtered_signal_g, fft_result_g,freq_g) = filter(data['g'])
(filtered_signal_h, fft_result_h,freq_h) = filter(data['h'])

min_freq = 50
max_freq = 70

#selected_freq_indices = np.where((freq >= min_freq) & (freq <= max_freq))[0]
plt.figure(figsize=(12, 8))
plt.subplot(428)
plt.subplot(421)
plt.plot(data['time'] ,filtered_signal_a)
plt.grid(True)
plt.subplot(422)
plt.plot(freq_a , np.abs(fft_result_a))
plt.grid(True)

plt.subplot(423)
plt.plot(data['time'] ,filtered_signal_b)
plt.grid(True)
plt.subplot(424)
plt.plot(freq_b , np.abs(fft_result_b))
plt.grid(True)

plt.subplot(425)
plt.plot(data['time'] ,filtered_signal_c)
plt.grid(True)
plt.subplot(426)
plt.plot(freq_c , np.abs(fft_result_c))
plt.grid(True)


plt.subplot(427)
plt.plot(data['time'] ,filtered_signal_d)
plt.grid(True)
plt.subplot(428)
plt.plot(freq_d , np.abs(fft_result_d))
plt.grid(True)

```

```python
def get_rms_acceleration(signal, frame_size, hop_length):
    sampling_freq = 200
    rms = []
    timestamps=[]
    for i in range(0, len(signal), hop_length):
        current_rms = np.sqrt(np.sum(signal[i:i+frame_size]**2)/frame_size)
        rms.append(current_rms)
        timestamps.append(i/sampling_freq)
    return rms, timestamps

frame_size = 50  # 프레임 크기 설정 하나의 구간에 점이 몇 개들어간느지
hop_length = 5 # 슬라이딩 윈도우의 이동 거리 설정    #시간축이랑 동시에 리스트 길이 조정하는게 필요 한구간이랑 다음구간이랑 겹치는 점 개수
# RMS 값을 추출
(rms_values_a,timestamps_a) = get_rms_acceleration(filtered_signal_a, frame_size, hop_length)
(rms_values_b,timestamps_b) = get_rms_acceleration(filtered_signal_b, frame_size, hop_length)
(rms_values_c,timestamps_c) = get_rms_acceleration(filtered_signal_c, frame_size, hop_length)
(rms_values_d,timestamps_d) = get_rms_acceleration(filtered_signal_d, frame_size, hop_length)
(rms_values_e,timestamps_e) = get_rms_acceleration(filtered_signal_e, frame_size, hop_length)
(rms_values_f,timestamps_f) = get_rms_acceleration(filtered_signal_f, frame_size, hop_length)
(rms_values_g,timestamps_g) = get_rms_acceleration(filtered_signal_g, frame_size, hop_length)
(rms_values_h,timestamps_h) = get_rms_acceleration(filtered_signal_h, frame_size, hop_length)


plt.figure(figsize=(12, 8))
plt.subplot(428)
plt.subplot(421)
plt.plot(timestamps_a,rms_values_a)
#plt.plot(data['time'] ,filtered_signal_a)
plt.grid(True)

plt.subplot(422)
plt.plot(timestamps_a,rms_values_b)
#plt.plot(data['time'] ,filtered_signal_b)
plt.grid(True)

plt.subplot(423)
plt.plot(timestamps_a,rms_values_c)
#plt.plot(data['time'] ,filtered_signal_c)
plt.grid(True)

plt.subplot(424)
plt.plot(timestamps_a,rms_values_d)
#plt.plot(data['time'] ,filtered_signal_d)
plt.grid(True)

plt.subplot(425)
plt.plot(timestamps_a,rms_values_e)
#plt.plot(data['time'] ,filtered_signal_e)
plt.grid(True)


plt.subplot(426)
plt.plot(timestamps_a,rms_values_f)
#plt.plot(data['time'] ,filtered_signal_f)
plt.grid(True)

plt.subplot(427)
plt.plot(timestamps_a,rms_values_g)
#plt.plot(data['time'] ,filtered_signal_g)
plt.grid(True)

plt.subplot(428)
plt.plot(timestamps_a,rms_values_h)
#plt.plot(data['time'] ,filtered_signal_h)
plt.grid(True)

plt.show
```
